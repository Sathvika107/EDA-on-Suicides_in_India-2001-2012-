# Exploraratory Data Analysis of Suicides in India


The analysis is done on the dataset Suicides in India 2001-2012.csv obtained from Kaggle.

```python
#importing required libraires
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
#reading the dataset file using panadas python interface
data=pd.read_csv("/Users/sathvikas/Desktop/suicide_dataset.csv")
```

```python
#dispaying the first and last 5 colums of the dataset 
data.head()
```

```python
data.tail()
```

```python
data.drop(data[(data['Total']==0)].index,inplace=True)# removing rows with values as null
```

```python
data.drop(data[(data.State=='Total (States)')|(data.State=='Total (All India)')|(data.State=='Total (Uts)')].index,inplace=True)
```

```python
data.State.value_counts() 
```

```python
#plotting a bar graph state wise suicides
state_count=data.groupby('State').sum()['Total'].sort_values(ascending=True).plot(kind='bar',title='Count of suicides in all states',figsize=(15,10),width=0.75,)
state_count.set_xlabel("States")
state_count.set_ylabel("No on suicides")

```

```python
#suicides genderwise
gender_count=data.groupby('Gender').sum()['Total'].plot(kind='bar',figsize=(6,6),title='Number of suicides')
gender_count.set_ylabel("No of suicides")
```

```python
#creating dataframes in accordance to the education status,means adopted,professional profile,social status
causes_df=data[data['Type_code']=='Causes']
edu_df=data[data['Type_code']=='Education_Status']
means=data[data['Type_code']=='Means_adopted']
profile=data[data['Type_code']=='Professional_Profile']
social=data[data['Type_code']=='Social_Status']


```

```python
#Age Suicide Frequency
data_age=data[data['Age_group']!='0-100+']

age = data_age.groupby('Age_group').sum()['Total'].plot(kind="bar",figsize=(8,5), title ="Age wise suicides frequency")
age.set_xlabel('\nAge Group')
age.set_ylabel('Counts\n')
sns.set_style('whitegrid')
sns.cubehelix_palette(8);
```

```python
#Plotting a line graph showing the number of suicides alongside the years
Year=data.groupby('Year').sum()['Total'].plot(kind='line',title='Count of suicides per year')
Year.set_xlabel('YEARS')
Year.set_ylabel('Count of suicides')
```

```python
Social = social.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(12,5), title ="Number of suicides versus social status ")
Social.set_xlabel("Social_status")
Social.set_ylabel("No of suicides")


```

```python
Prof= profile.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(12,5), title ="Number of suicies versus Profession ")
Prof.set_xlabel("Profession")
Prof.set_ylabel("No of suicides")

```

```python
Method = means.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(12,5), title ="Number of suicides versus means adopted")
Method.set_xlabel("Means adopted")
Method.set_ylabel("No of suicides")

```

```python
Educ =edu_df.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(8,5), title ="Number of suicides versus education")
Educ.set_xlabel("Education")
Educ.set_ylabel("No of suicides")

```

```python
Causess = causes_df.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(12,8), title ="Number of suicides versus causes for suicide") 
Causess.set_xlabel("Causes")
Causess.set_ylabel("No of suicides")
```

```python
#creating seperate dataframes of the data seperating them according to their age
age_df1=causes_df[causes_df['Age_group']=='0-14']
age_df2=causes_df[causes_df['Age_group']=='15-29']
age_df3=causes_df[causes_df['Age_group']=='30-44']
age_df4=causes_df[causes_df['Age_group']=='45-59']
age_df5=causes_df[causes_df['Age_group']=='60+']

```

```python
#plotting the number of suicides age group wise
AGE = age_df1.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(12,9), title ="Number of suicides in the age group 0-14 ")
AGE.set_xlabel("Causes")
AGE.set_ylabel("No of suicides")
```

```python
AGE1=age_df2.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(12,9), title ="Number of suicides in the age group 15-29")
AGE1.set_xlabel("Causes")
AGE1.set_ylabel("No of suicides")
```

```python
AGE2=age_df3.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(12,9), title ="Number of suicides in the age group 30-44")
AGE2.set_xlabel("Causes")
AGE2.set_ylabel("No of suicides")
```

```python
AGE3=age_df4.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(12,9), title ="Number of suicides in the age group 45-59")
AGE3.set_xlabel("Causes")
AGE3.set_ylabel("No of suicides")
```

```python
AGE4=age_df5.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(12,9), title ="Number of suicides in the age group 60+")
AGE4.set_xlabel("Causes")
AGE4.set_ylabel("No of suicides")
```

# Conclusion from the analysis done

 - It can be seen that the number of suicides in the state Maharashtra is the highest in the period of 2001 to 2012.
 - The number of suicides among the men is very high comapred to women.
 - Individuals between age 15-29 have comitted suicide the most.
 - Hanging is the most used means of committing suicide
